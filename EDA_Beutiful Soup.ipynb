```python
from bs4 import BeautifulSoup
```

```python
page = open("03.zerobase.html","r").read()
print(page)
```

```python
soup = BeautifulSoup(page,"html.parser")
print(soup.prettify)
```

```python
#head 태그 확인
soup.head
```

```python
#body 태그 확인
soup.body
```

```python
#p태그 확인
# 처음 발견한 p 태그 알려줌
soup.p
```

```python
# 한 개 찾을때 > 여러개 find_all()
soup.find("p")
```

```python
soup.find("p",class_= "inner-text first-item")
```

```python
soup.find("p",{"class":"outer-text first-item"})
```

```python
#html 코드 지우기
soup.find("p",{"class":"outer-text first-item"}).text.strip()
```

```python
# 다중조건
soup.find("p",{"class":"inner-text first-item","id":"first"})
```

```python
# find_all() : 여러 개의태그를 반환
# list로 반환
soup.find_all("p")
```

```python
# 특정 태그 확인
soup.find_all(class_="outer-text")
```

```python
# 특정 태그 확인2
soup.find_all(id="pw-link")[0].text
```

```python
soup.find_all("p",class_="inner-text first-item")
```

```python
len(soup.find_all("p"))
print(soup.find_all("p")[0].text)
print(soup.find_all("p")[1].string)
print(soup.find_all("p")[1].get_text())
```

```python
# p 태그 리스트에서 텍스트 속성만 출력
for each_tag in soup.find_all("p"):
    print("=" * 50)
    print(each_tag.text)
```

```python
# a 태그에서 href 속성값에 있는 값 추출
links = soup.find_all("a")
links[0].get("href"),links[1]["href"]
links
```

```python
for each in links:
    href = each.get("href") #each["href"]
    text = each.get_text()
    print(text + "=>" + href)    
```
